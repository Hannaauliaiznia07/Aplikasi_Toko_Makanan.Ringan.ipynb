```python
"""
ðŸ§ª PENGUJIAN & EVALUASI - TUGAS 4
Aplikasi: Toko Makanan Ringan
-----------------------------------
Pengujian dilakukan untuk memastikan:
- Semua fitur utama berjalan sesuai harapan
- Input valid dan tidak valid ditangani dengan baik
"""

# Import fungsi dari program utama jika terpisah,
# atau copy paste program langsung di bawah bagian pengujian ini jika digabung

# Versi singkat kode utama (bisa juga diimpor dari file utama)
menu = [
    {"nama": "Keripik Singkong", "harga": 8000},
    {"nama": "Kacang Atom", "harga": 10000},
    {"nama": "Coklat Batang", "harga": 12000}
]

def lihat_menu():
    print("\n=== Daftar Makanan Ringan ===")
    for i, item in enumerate(menu, 1):
        print(f"{i}. {item['nama']} - Rp{item['harga']:,}")
    print()

def tambah_makanan(nama, harga):
    try:
        harga = int(harga)
        menu.append({"nama": nama, "harga": harga})
        print(f"Makanan '{nama}' berhasil ditambahkan.")
    except ValueError:
        print("ERROR: Harga harus berupa angka!")

def pesan_makanan(pilihan, jumlah):
    try:
        pilihan = int(pilihan)
        jumlah = int(jumlah)
        if 1 <= pilihan <= len(menu):
            item = menu[pilihan - 1]
            total = item['harga'] * jumlah
            print(f"Anda memesan {jumlah} x {item['nama']}")
            print(f"Total harga: Rp{total:,}")
        else:
            print("ERROR: Nomor makanan tidak tersedia.")
    except ValueError:
        print("ERROR: Masukkan angka yang valid!")

# -------------------------------
# ðŸ§ª PENGUJIAN OTOMATIS SEDERHANA
# -------------------------------

print(">>> TEST 1: Lihat Daftar Makanan")
lihat_menu()

print("\n>>> TEST 2: Tambah Makanan Baru (Valid)")
tambah_makanan("Basreng Pedas", 9000)
lihat_menu()

print("\n>>> TEST 3: Tambah Makanan Baru (Invalid Harga)")
tambah_makanan("Permen Kopi", "abc")

print("\n>>> TEST 4: Pesan Makanan (Valid)")
pesan_makanan(1, 2)

print("\n>>> TEST 5: Pesan Makanan (Nomor Salah)")
pesan_makanan(99, 1)

print("\n>>> TEST 6: Pesan Makanan (Jumlah Salah)")
pesan_makanan(1, "tiga")

print("\n>>> TEST 7: Program Selesai â€” Semua tes dijalankan.")
```
